```python
# importing libraries
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.dummy import DummyRegressor
import pandas as pd
import html
import time
import csv
# from scrapy.downloadermiddlewares.retry import RetryMiddleware
# from scrapy_random_proxies.middlewares import RandomProxyMiddleware
import requests
from bs4 import BeautifulSoup
import random
from requests_ip_rotator import ApiGateway, EXTRA_REGIONS

import os
```

```python
friends_problemset_50_df = pd.read_csv('/content/friends_problemset_50_df_without_source_code.csv')
friends_problemset_50_df
```

```python
top_50_friends = friends_problemset_50_df.groupby('Who')['#'].count().reset_index().sort_values('#', ascending= False)[:50]
top_50_friends
```

```python
header = ['#','When', 'Who', 'Problem_name', 'Number', 'Character']
top_friends_problemset_df = pd.DataFrame(columns=header)
for index, name in top_50_friends['Who'].iteritems():
  print(name)
  # Send GET request to the page
  url = 'https://codeforces.com/submissions/{}'

  friend_name = name

  # Make a GET request to the submission URL with the contest ID and submission ID
  response = requests.get(url.format(friend_name))

  # Check if the request was successful
  if response.status_code == 200:
      # Parse the HTML content using BeautifulSoup
      soup = BeautifulSoup(response.content, 'html.parser')

  pagination_div = soup.find('div', id= 'body').find_all('div',class_= 'pagination')
  if pagination_div is not None and len(pagination_div)>1:
      pages = pagination_div[1].find_all('a')
      total_pages = int(pages[-2].text.strip())
      print(total_pages)
  else:
      total_pages = 1
  # Find the submissions table
  data = []

  for page in range(1, total_pages + 1):
    page_url = url + '/page/' + str(page)
    print(page_url.format(friend_name))
    response = requests.get(page_url.format(friend_name))
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table', {'class': 'status-frame-datatable'})
    if table is None:
        print("Error: Submissions table not found")
        exit()

    for row in table.find_all('tr'):
      cols = row.find_all('td')
      if len(cols) > 0:
        if cols[5].text.strip()=='Accepted':
          if (cols[0].find('span') and 'hiddenSource' in cols[0].find('span').get('class') ) or (cols[0].find('a') and 'contest' in cols[0].find('a')['href']):
            when = cols[1].text.strip()
            num = row.find_all('td', class_="status-small")[1].find('a')['href'].split('/')[2]
            Problem_name = cols[3].text.split('-')[1].strip()
            Character = cols[3].text.split('-')[0].strip()
            submission = cols[0].text.strip()
            data.append([submission, when, friend_name, Problem_name, num, Character])

  # Create a pandas DataFrame from the rows
  friends_problem_df = pd.DataFrame(data, columns=header)
  top_friends_problemset_df = pd.concat([top_friends_problemset_df,friends_problem_df], axis=0)
```

```python
top_friends_problemset_df['#'] = top_friends_problemset_df['#'].astype('int64')
```

```python
mask = ~(top_friends_problemset_df['#'].isin(friends_problemset_50_df['#'].values) &
         top_friends_problemset_df['When'].isin(friends_problemset_50_df['When'].values) &
         top_friends_problemset_df['Who'].isin(friends_problemset_50_df['Who'].values)
         )


# Filter top_friends_problemset_df using the boolean mask
filtered_df = top_friends_problemset_df[mask]

# Display the filtered DataFrame
filtered_df[filtered_df['Who']=='kinannotfound']
```
