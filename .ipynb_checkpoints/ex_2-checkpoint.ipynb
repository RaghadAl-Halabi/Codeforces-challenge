```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
left = ctrl.Antecedent(np.arange(0, 256, 1), 'left')
right = ctrl.Antecedent(np.arange(0, 256, 1), 'right')
front = ctrl.Antecedent(np.arange(0, 256, 1), 'front')
back = ctrl.Antecedent(np.arange(0, 256, 1), 'back')
left_motor = ctrl.Consequent(np.arange(-100, 101, 1), 'left_motor')
right_motor = ctrl.Consequent(np.arange(-100, 101, 1), 'right_motor')
```

```python
left['close'] = fuzz.trapmf(left.universe, [0, 0, 30, 100])
left['medium'] = fuzz.trapmf(left.universe, [50, 100, 150, 200])
left['far'] = fuzz.trapmf(left.universe, [150, 225, 255, 255])

left.view()
```

```python
right['close'] = fuzz.trapmf(left.universe, [0, 0, 30, 100])
right['medium'] = fuzz.trapmf(right.universe, [50, 100, 150, 200])
right['far'] = fuzz.trapmf(right.universe, [150, 225, 255, 255])

right.view()
```

```python
front['close'] = fuzz.trapmf(left.universe, [0, 0, 30, 100])
front['medium'] = fuzz.trapmf(front.universe, [50, 100, 150, 200])
front['far'] = fuzz.trapmf(front.universe, [150, 225, 255, 255])

front.view()
```

```python
back['close'] = fuzz.trapmf(left.universe, [0, 0, 30, 100])
back['medium'] = fuzz.trapmf(back.universe, [50, 100, 150, 200])
back['far'] = fuzz.trapmf(back.universe, [150, 225, 255, 255])

back.view()
```

```python
left_motor['stop'] = fuzz.trapmf(left_motor.universe, [-5,-5, 5, 5])
left_motor['low_front'] = fuzz.trimf(left_motor.universe, [0, 40, 80])
left_motor['high_front'] = fuzz.trimf(left_motor.universe, [70, 100, 100])
left_motor['low_back'] = fuzz.trimf(left_motor.universe, [-80, -40, 0])
left_motor['high_back'] = fuzz.trimf(left_motor.universe, [-100, -100, -70])

left_motor.view()
```

```python
right_motor['stop'] = fuzz.trapmf(right_motor.universe, [-5,-5, 5, 5])
right_motor['low_front'] = fuzz.trimf(right_motor.universe, [0, 40, 80])
right_motor['high_front'] = fuzz.trimf(right_motor.universe, [70, 100, 100])
right_motor['low_back'] = fuzz.trimf(right_motor.universe, [-80, -40, 0])
right_motor['high_back'] = fuzz.trimf(right_motor.universe, [-100, -100, -70])

right_motor.view()
```

```python

rule1 = ctrl.Rule(back['medium'] , (left_motor['low_back'], right_motor['low_back'] ))
rule2 = ctrl.Rule(left['close'] & right['close'] & front['close'] & back['close'], (left_motor['stop'], right_motor['stop'] ))
rule3 = ctrl.Rule(left['close'] & right['close'] & front['close'] & back['far'], (left_motor['high_back'], right_motor['high_back'] ))
rule4 = ctrl.Rule(left['close'] & right['close'] & front['close'] & back['medium'], (left_motor['low_back'], right_motor['low_back'] ))
rule5 = ctrl.Rule(front['close'] & back['close']& right['close'], (left_motor['high_back'], right_motor['stop'] ))
rule6 = ctrl.Rule(front['close'] & back['close']& left['close'], (left_motor['stop'], right_motor['high_back'] ))
rule7 = ctrl.Rule(front['far'] , (left_motor['high_front'], right_motor['high_front'] ))
rule8 = ctrl.Rule(front['medium'] , (left_motor['low_front'], right_motor['low_front'] ))
rule9 = ctrl.Rule(front['medium'] & back['far'] & left['close'] & right['close'], (left_motor['high_back'], right_motor['high_back'] ))
```

```python
control_system = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])
```

```python
simulation = ctrl.ControlSystemSimulation(control_system)
```

```python
simulation.input['left'] = 30
simulation.input['right'] = 30
simulation.input['front'] = 150
simulation.input['back'] = 30
simulation.compute()
```

```python
print(simulation.output['left_motor'])
print(simulation.output['right_motor'])
left_motor.view(sim=simulation)
right_motor.view(sim=simulation)
```
